```python
data = pd.read_csv('Fremont_Bridge_Bicycle_Counter.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
data.columns = ['Total', 'East', 'West']
```

```python
data.head()
```

```python
data.dropna().describe()
```

```python
%matplotlib inline
import seaborn; seaborn.set()

data.plot()
```

```python
import matplotlib.pyplot as plt

weekly = data.resample('W').sum()
weekly.plot(style=[':', '--', '-'])
plt.ylabel('Weekly bicycle count')
```

```python
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=[':', '--', '-'])
plt.ylabel('mean hourly count')

```

```python
by_time = data.groupby(data.index.time).mean()
houly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=houly_ticks, style=[':', '--', '-'])

# 시간대별 통행량
```

```python
#요일에 따른 통행량

by_weekday = data.groupby(data.index.dayofweek).mean()
by_weekday.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
by_weekday.plot(style=[':', '--', '-'])
```

```python
# 복합적인 groupby를 사용해 주중과 주말의 시간대별 추이를 살펴보자.

weekend = np.where(data.index.weekday < 5, 'Weekday', 'Weekend')
by_time = data.groupby([weekend, data.index.time]).mean()

fig, ax = plt.subplots(1, 2, figsize=(14,5))
by_time.loc['Weekday'].plot(ax=ax[0], title='Weekdays', xticks=houly_ticks, style=[':', '--', '-'])
by_time.loc['Weekend'].plot(ax=ax[1], title='Weekends', xticks=houly_ticks, style=[':', '--', '-'])
```
